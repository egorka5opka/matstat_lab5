Воспользуемся One-way ANOVA. 

положим $y = Sepal.Length \cdot Sepal.Width + Petal.Length\cdot Petal.Width$

$y_{i, j} = \mu_i + \epsilon_{, j}$, где $\epsilon_{i, j} \sim N(0, \sigma^2)$

$\mu_i$ --- среднее значение в группе $i$
$e_{i,j}$ --- шум

Гипотезы:
$H_0$: $\mu_1 = \mu_2 = \mu_3$
$H_1$: $\lnot H_0$

$S_B^2 = \sum_{i=1}^I J_i (\bar y_i - \bar y)^2$, где $\bar y_i = \frac{1}{J_i} \sum_{j=1}^{J_i}y_{i,j}$
